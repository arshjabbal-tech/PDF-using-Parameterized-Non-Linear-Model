```python
import numpy as np
import pandas as pd

# Load dataset
df = pd.read_csv("/content/data.csv")

x = df["no2"]

r = 102303561
a_r = 0.05 * (r % 7)
b_r = 0.3 * ((r % 5) + 1)

z = x + a_r * np.sin(b_r * x)

mu = np.mean(z)
variance = np.var(z)

lambda_ = 1 / (2 * variance)
c = 1 / np.sqrt(2 * np.pi * variance)

print(mu, lambda_, c)
```
